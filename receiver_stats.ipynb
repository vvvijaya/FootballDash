```python
df = pd.read_parquet('2023_pbp.parquet') #pass 
df.query("play_type=='pass'")[['pass_touchdown','play_type','air_yards','yards_after_catch','yards_gained','touchdown','receiver_player_id','receiver_player_name','passer']].groupby(['passer','receiver_player_name']).count().reset_index()
```

```python
df.query("play_type=='pass'")[['passer', 'receiver_player_name']].groupby(['passer', 'receiver_player_name']).size().reset_index()
```

```python
df_pass_td = df.query("play_type=='pass'")[['passer', 'receiver_player_name', 'pass_touchdown']].groupby(['passer', 'receiver_player_name']).sum().reset_index()
df_pass_td['total_pass'] = df.query("play_type=='pass'")[['passer', 'receiver_player_name']].groupby(['passer', 'receiver_player_name']).size().reset_index(name='total')[['total']]
df_pass_td['td_to_total_ratio'] = df_pass_td['pass_touchdown']/df_pass_td['total_pass']
df_pass_td
```
